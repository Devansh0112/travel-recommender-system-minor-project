```python
import pandas as pd
import numpy as np
df = pd.read_csv('cities.csv')
df.head()
```

```python
df1 = df[['name_of_city','population_total','effective_literacy_rate_total','total_graduates']].copy()
df1.head(10)
```

```python
df1['sample_mean'] = df1.mean(axis=1)
df1.head()
```

```python
df1['sample_mean'] = df1['sample_mean']/df1['sample_mean'].max()
df1.head()
```

```python
a = ['Abohar ','Delhi ','Dehri ']
b = list()
for i in a:
    b.append(df1.loc[df1['name_of_city'] == i , 'sample_mean'].iloc[0])
    
print(b)
```

```python
b = np.mean(b)
b
```

```python
filter1 = df1['sample_mean'] > b-0.1
filter2 = df1['sample_mean'] < b+0.1

d = df1['name_of_city'].where(filter1 & filter2)
d = d.dropna()
d
e = list(d)
e
```
